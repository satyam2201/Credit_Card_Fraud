Under Sampling 

Build a sample dataset contaiing similar distribution of normal transction and fraudlent transction

Number of Fraudelent DataSet = 146

```python
legit_sample = legit.sample(n=146)
```

```python
# concatinating two data frames
```

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
x = new_dataset.drop(columns='Class',axis=1)
y = new_dataset['Class']
```

Spliting the data into Features and Traget

```python
x
```

```python
y
```

Split the data into traing data and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training


Logistic Training

```python
model = LogisticRegression()
```

```python
# traing the Logistic Regression model with traing data
model.fit(x_train,y_train)
```

Accuracy Score


```python
#accuracy score on traing data
x_train_prediction = model.predict(x_train)
training_data_acuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print("acccuracy on traing data ----> ",training_data_acuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print("acccuracy on test data ----> ",test_data_accuracy)
```
